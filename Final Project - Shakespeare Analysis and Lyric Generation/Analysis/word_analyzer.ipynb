```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import re

def create_graph(paragraph):
    G = nx.DiGraph()
    words = re.findall(r'\w+', paragraph)
    words = [word.lower() for word in words]
    for i in range(len(words)):
        G.add_node(words[i])
        if i < len(words)-1:
            G.add_edge(words[i], words[i+1])
    return G

def draw_graph(G):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True)
    plt.show()


#paragraph = "The quick brown fox jumps over the lazy dog, and after he did that the dog bit the fox in the leg"
paragraph = "Shall I compare thee to a summers day?\
Thou art more lovely and more temperate:\
Rough winds do shake the darling buds of May,\
And summers lease hath all too short a date;\
Sometime too hot the eye of heaven shines,\
And often is his gold complexion dimm'd;\
And every fair from fair sometime declines,\
By chance or natures changing course untrimm'd;\
But thy eternal summer shall not fade,\
Nor lose possession of that fair thou owst;\
Nor shall death brag thou wanderst in his shade,\
When in eternal lines to time thou growst:\
So long as men can breathe or eyes can see,\
So long lives this, and this gives life to thee."

G = create_graph(paragraph)
draw_graph(G)
```

```python
diam = nx.diameter(G)
print("N =",len(G.nodes()))
print("Diameter =", diam)
```
