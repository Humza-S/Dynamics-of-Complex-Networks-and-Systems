```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import re
import pickle
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
def load_pkl(file):
    return pickle.load(open(file, "rb" ))
```

```python
shakespeare_poems = load_pkl('shakespeare_poems.p')
```

```python
poems = list(shakespeare_poems.values())
```

```python
pattern = r"\b\w+(?:[‘’'-]\w+|\w+'\w+)?\b"

tokenizer = RegexpTokenizer(pattern)
```

```python
stop_words = set(stopwords.words('english'))
stop_words.update(['hadst', 'hath', 'hast', 'wast', 'wert', 'art', 'didst', 'dost', 'doth', 'canst', 'mayst', 'may', 'must', 'shalt', 'shall', 'shouldst', 'thy', 'thou', 'thine', 'wilt', 'will', 'wouldst', 'thee','ye', 'i’ll', 'tis', 'yet', 'us', 'erst', 'let', 'there\'s', 'lo', 'that’s', 'whenas', 'th', 'he’ll'])
```

```python
lemma_words = {}
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
set(stopwords.words('english'))
```

```python
for lyrics in poems:
    for lyric in lyrics:
        print(lyric)
        words = tokenizer.tokenize(lyric.lower())
        print(words)
        filtered_stopwords = [word for word in words if word not in stop_words]
        print(filtered_stopwords)
        break
```

```python
for lyrics in poems:
    for lyric in lyrics:
        
        if len(lyric) == 0:
            continue
        
        words = tokenizer.tokenize(lyric.lower())
        filtered_stopwords = [word for word in words if word not in stop_words]
        
        for word in filtered_stopwords:
            lemmetized_word = lemmatizer.lemmatize(word)
            if lemmetized_word not in lemma_words.keys():
                lemma_words[lemmetized_word] = [word]
            elif lemmetized_word in lemma_words.keys() and word not in lemma_words[lemmetized_word]:
                lemma_words[lemmetized_word] = lemma_words[lemmetized_word] + [word] 
        
```

```python
biggest_lemma = 0
big_k = ''
big_v = []
for k,v in lemma_words.items():
    if len(v) > biggest_lemma:
        biggest_lemma = len(v)
        big_k = k
        big_v = v
```

```python
from nltk.stem.porter import *
```

```python
stemmer = PorterStemmer()
```

```python
stemmer_words = {}
```

```python
SG = nx.DiGraph()
```

```python
for lyrics in poems:
    for lyric in lyrics:
        
        if len(lyric) == 0:
            continue
        
        words = tokenizer.tokenize(lyric.lower())
        filtered_stopwords = [word for word in words if word not in stop_words]
        
        for word in filtered_stopwords:
            stemmed_word = stemmer.stem(word)
            if stemmed_word not in stemmer_words.keys():
                stemmer_words[stemmed_word] = [word]
            elif stemmed_word in stemmer_words.keys() and word not in stemmer_words[stemmed_word]:
                stemmer_words[stemmed_word] = stemmer_words[stemmed_word] + [word] 
                
            SG.add_node(stemmed_word)
            
        for i in range(len(filtered_stopwords)-1):
            source = stemmer.stem(filtered_stopwords[i])
            target = stemmer.stem(filtered_stopwords[i+1])

            if SG.has_edge(source, target):
                SG[source][target]['weight'] += 1
            else:
                SG.add_edge(source, target, weight=1)
        
```

```python
nx.write_gml(SG, "shakespeare_stemmed.gml")
```
